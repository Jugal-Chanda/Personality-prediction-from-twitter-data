## Calculating DF for all words

```python
N = len(document)
DF = {}

for i in range(N):
    tokens = word_tokenize(text[i])
    for w in tokens:
        try:
            DF[w].add(i)
        except:
            DF[w] = {i}
for i in DF:
    DF[i] = len(DF[i])
```

```python
def doc_freq(word):
    c = 0
    try:
        c = DF[word]
    except:
        pass
    return c
```

### Calculating TF-IDF for body, we will consider this as the actual tf-idf as we will add the title weight to this.

```python
doc = 0

tf_idf = {}

for i in range(N):
    
    tokens = word_tokenize(text[i])
    
    counter = Counter(tokens)
    words_count = len(tokens)
    
    for token in np.unique(tokens):
        
        tf = counter[token]/words_count
        dof = doc_freq(token)
        idf = np.log((N+1)/(dof+1))
        
        tf_idf[doc, token] = tf*idf
    doc += 1
```

```python
type(tf_idf)
tf_idf
```

```python
tfidf_individual_I = {}
tfidf_individual_E = {}
tfidf_individual_N = {}
tfidf_individual_S = {}
tfidf_individual_T = {}
tfidf_individual_F = {}
tfidf_individual_J = {}
tfidf_individual_P = {}
for key in tf_idf:
    if key[0] == 0:
        tfidf_individual_I[key[1]] = round(tf_idf[key],10)
    elif key[0] == 1:
        tfidf_individual_E[key[1]] = round(tf_idf[key],10)
    elif key[0] == 2:
        tfidf_individual_N[key[1]] = round(tf_idf[key],10)
    elif key[0] == 3:
        tfidf_individual_S[key[1]] = round(tf_idf[key],10)
    elif key[0] == 4:
        tfidf_individual_T[key[1]] = round(tf_idf[key],10)
    elif key[0] == 5:
        tfidf_individual_F[key[1]] = round(tf_idf[key],10)
    elif key[0] == 6:
        tfidf_individual_J[key[1]] = round(tf_idf[key],10)
    elif key[0] == 7:
        tfidf_individual_P[key[1]] = round(tf_idf[key],10)
```
